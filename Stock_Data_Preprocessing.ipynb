```python
import pandas as pd
```

```python
apple = pd.read_csv('/Users/asmita/Desktop/StockMarketDashboard/StockMarketDataset/stocks/AAPL.csv')
facebook = pd.read_csv('/Users/asmita/Desktop/StockMarketDashboard/StockMarketDataset/stocks/FB.csv')
google = pd.read_csv('/Users/asmita/Desktop/StockMarketDashboard/StockMarketDataset/stocks/GOOGL.csv')
nvidia = pd.read_csv('/Users/asmita/Desktop/StockMarketDashboard/StockMarketDataset/stocks/NVDA.csv')
tesla = pd.read_csv('/Users/asmita/Desktop/StockMarketDashboard/StockMarketDataset/stocks/TSLA.csv')
twitter = pd.read_csv('/Users/asmita/Desktop/StockMarketDashboard/StockMarketDataset/stocks/TWTR.csv')
```

```python
apple.head()
```

```python
dfs = [apple, facebook, google, nvidia, tesla, twitter]
```

```python
for df in dfs:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA200'] = df.Close.rolling(200).mean()
```

```python
apple.head(200)
```

```python
for df in dfs:
    df['Previous day close price'] = df.Close.shift(1)

for df in dfs:
    df['Percent change in price'] = df.Close.pct_change()

for df in dfs:
    df['Percent change in price'] = df.Close.pct_change()

for df in dfs:
    df['Previous day volume'] = df.Volume.shift(1)

for df in dfs:
    df['Change in volume'] = df['Volume'] - df['Previous day volume']

for df in dfs:
    df['Percent change in volume'] = df.Volume.pct_change()
```
